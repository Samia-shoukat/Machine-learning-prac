```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("encoding.csv")
dataset.head(5)
```

```python
dataset.isnull().sum()
```

```python
dataset["Salary"].fillna(dataset["Salary"].mean(),inplace=True)
```

```python
sns.distplot(dataset["Application_Income"])
plt.show()
```

```python
dataset.describe()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
ss.fit(dataset[["Application_Income"]])
```

```python
dataset["Application_ss"]= pd.DataFrame(ss.transform(dataset[["Application_Income"]]), column = ["x"])
```

```python
dataset.head(5)
```

```python
dataset.describe()
```

```python
plt.subplot(1,2,1)
sns.distplot(dataset["Application_ss"])
plt.subplot(1,2,2)
sns.distplot(dataset["Application_Income"])
plt.show()
```
