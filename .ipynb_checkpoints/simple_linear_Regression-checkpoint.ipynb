```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
dataset = pd.read_csv(r"E:\Machine Learning\CGPA_Package.csv")
dataset.head(5)
```

```python
sns.scatterplot(x="cgpa", y="package", data=dataset)
```

```python
dataset.isnull().sum()
```

```python
x = dataset[["cgpa"]]
y = dataset["package"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
# y = m*x+c
```

```python
lr.score(x_test,y_test)*100
```

```python
lr.predict([[6.89]])
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
0.46857662*6.89+0.1584659795716492
```

```python
y_prd = lr.predict(x)
```

```python
sns.scatterplot(x="cgpa", y="package", data=dataset, label="org")
plt.plot(dataset["cgpa"],y_prd, c="red")
plt.show()
```
